```python
def quick(A,l,r):
  if(r-l<=1):
    return A
  
  yel=l+1
  for gre in range(l+1,r):
    if(A[gre]<=A[l]):
      swap(A,yel,gre)
      yel=yel+1
      
      
  swap(A,l,yel-1)
  quick(A,l,yel)
  quick(A,yel,r)

  return A

def swap(A,top,bot):
  temp=A[top]
  A[top]=A[bot]
  A[bot]=temp
  
  
def merge(A,B,C):
  n=len(A)
  m=len(B)
  i=0
  j=0
  while(i<n and j<m):
    if(A[i]<=B[j]):
      C.append(A[i])
      i+=1
    else:
      C.append(B[j])
      j+=1
     
      
  
  while(i>=n):
      C.append(B[j])
      j+=1
      if(j==m):
        return C
     
      
  while(j>=m):
      C.append(A[i])
      i+=1
      if(i==n):
        return C
      
      
x=list(map(int,input().split()))
u=[]
for i in range(5):
  u.append(list(map(int,input().split())))

s=[]
for i in range(len(u)):
  s.append(quick(u[i],0,len(u[i])))
   
k=[]
for i in range(len(s)):
  k=merge(k,s[i],[]) 

p=[]
i=0
while(i<=len(k)-3):
  flag=0
  j=i
  while(j<i+3):
    if(k[i]==k[j]):
      flag+=1
    j+=1
      
    
    
  if(flag>=3):
    p.append(k[i])
  i+=flag   
print(len(p))  
```
